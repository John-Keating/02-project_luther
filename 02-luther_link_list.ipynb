### 1. Scrape box office mojo list of pages with foriegn language films

```python
## import libraries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
# import the main page with requests

url = 'http://www.boxofficemojo.com/genres/chart/?id=foreign.htm'

response = requests.get(url)
```

```python
## return unicode

page = response.text #at this point I am getting the 
```

```python
soup = BeautifulSoup(page)
```

```python
print soup.find_all('b')[2]
```

```python
link_soup = soup.find_all('b')[2]
```

```python
for link in link_soup.find_all('a'):
    print link
```

```python
link_list = []
for link in link_soup.find_all('a'):
    link_list.append(str(link))

print type(link_list[0])
print link_list
```

```python
base_url = 'http://www.boxofficemojo.com'
url_list = []

for url in link_list:
    x = url.replace('<a href="', 'http://www.boxofficemojo.com').partition('">')
    y = x[0].replace('amp;', '') # I should go a
    url_list.append(y)

print url_list
```

```python
## append the main page the list of requests
url_list = [url] + url_list

## pickle the list
import pickle

with open('url_list.pk1', 'w') as picklefile:
    pickle.dump(url_list, picklefile)
```
